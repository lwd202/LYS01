## 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

    ◾ 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

        ◾ count: Counter 인스턴스가 지니는 초기 정수 값
        ◾ step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

    ◾ 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

    ◾ 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

        >>> c.incr()

        >>> print(c) 
        [Count (step: 1)] 11 
        >>> print(d) 
        [Count (step: 2)] 12
        
    ◾ 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

        >>> c() 
        >>> d() 
        >>> print(c) 
        [Count (step: 1)] 12 
        >>> print(d) 
        [Count (step: 2)] 14 

    ◾ 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> c = c + 5 
        >>> d = d - 5 
        >>> print(c) 
        [Count (step: 1)] 17 
        >>> print(d) 
        [Count (step: 2)] 9 

    ◾ 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> print(c) > 10 
        True 
        >>> print(d) > 10 
        False 
        >>> print(c) < 10 
        False 
        >>> print(d) < 10 
        True 
        >>> print(c) == 17 
        True 
        >>> print(d) != 9 
        False 

```python
class Counter():
    def __init__(self, count, step = 1): #생성자
        self.count = count
        self.step = step
    def __str__(self): #출력 메소드
        return str('[Count (step : ' + str(self.step) + ')] ' + str(self.count))
    def incr(self): #count를 step만큼 증가
        self.count = self.count + self.step
    def __call__(self): #객체 호출
        self.incr()
    #산술 연산자
    def __add__(self, other):
        return Counter(self.count + other, self.step)
    def __sub__(self, other):
        return Counter(self.count - other, self.step)
    # 관계 연산자
    def __gt__(self, y):
        return self.count > y
    def __ge__(self, y):
        return self.count >= y    
    def __lt__(self, y):
        return self.count < y    
    def __le__(self, y):
        return self.count <= y        
    def __eq__(self, y):
        return self.count == y
    def __ne__(self, y):
        return self.count != y
    
#1
c = Counter(10)
d = Counter(10, 2)

#2
print(c)
print(d)
print()

#3
c.incr()
d.incr()
print(c)
print(d)
print()

#4
c()
d()
print(c)
print(d)
print()

#5
c = c + 5
d = d - 5
print(c)
print(d)
print()

#6
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

```python
class MySet(list):
    #생성자, list를 인자로 받아 각각의 원소를 추가함.
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    #출력형태를 정해줌.
    def __str__(self):
        result = "MySet: {"
        for e in self: #각각의 원소에 대해 result에 문자열로 추가
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}" #마지막 ,를 제거
        return result

    @staticmethod    #정적 메소드, 중복검사를 수행하는 메소드
    def eliminate_duplicate(l):
        s = []
        for e in l: #리스트 l에 대해 각각의 원소 e가
            if e not in s: #s에 존재하지 않으면
                s.append(e) #s에 e를 추가
        l[:] = [] #l을 초기화
        for e in s: #s의 원소 e를
            l.append(e) #l에 추가

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
```

## 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

    ◾ 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
    ◾ 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

        >>> u = s | t 
        >>> print(u) 
        MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

    ◾ 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

        >>> u = s & t 
        >>> print(u) 
        MySet: {2, 3} 

    ◾ 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

        >>> s = MySet([1, 2, 3]) 
        >>> t = MySet([3, 4, 5]) 
        >>> u = s - t 
        >>> print(u) 
        MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    #요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    def __or__(self, other):
        for each in other: #other 집합의 원소가
            if (each not in self): #self에 없으면
                self.append(each) #self에 원소를 추가함
        self.sort() #오름차순 정렬
        return MySet(self)
    #요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    def __and__(self, other):
        result = []
        for each in other: #other 집합의 원소가
            if (each in self): #self에 있고
                if (each not in result): #result에 없으면
                    result.append(each) #result에 원소를 추가함
        self.sort() #오름차순 정렬
        return MySet(result)
    #요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    def __sub__(self, other):
        for each in other: #other 집합의 원소가
            if (each in self): #self에 있으면
                self.remove(each) #self에서 그 원소를 삭제함
        self.sort() #오름차순 정렬
        return MySet(self)
    
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    
u = s | t
print(u)

s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])

u = s & t
print(u)

s = MySet([1, 2, 3]) 
t = MySet([3, 4, 5]) 
u = s - t
print(u) 
```

        * 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.

```python
from urllib import request

def PrintHtml(url):
    f = request.urlopen(url)
    source_bytes = f.read()
    source = source_bytes.decode('utf-8')
    f.close()
    # \n, \t 제거
    temp = source.replace('\n', '')
    temp = temp.replace('\t', '')
    temp = temp.replace('\r', '')
    temp = temp.replace('>', '>$')

    while(temp.find('<script') != -1):
        start =  temp.find('<script')
        end = temp.find('/script>')
        temp = temp[:start] + temp[end+8:]
        
    while(temp.find('<style') != -1):
        start =  temp.find('<style')
        end = temp.find('/style>')
        temp = temp[:start] + temp[end+7:]

    temp = temp.split('<')    
    result = []
    
    x = 0
    
    while x < len(temp):
        z = temp[x]
        if(('src=' in z) or ('href=' in z)):
            temp.remove(z)
        else:
            x = x + 1
    x = 0

    while x < len(temp):
        temp[x] = temp[x].strip()
        if(temp[x].find('$') != -1):
            temp2 = temp[x].split('$')
            y = 0
            while y < len(temp2):
                temp2[y].strip()
                if((temp2[y].find('>') == -1) and (temp2[y] != '')):
                    result.append(temp2[y])
                y = y + 1
        x = x + 1


    result = ' '.join(result)
    result = result.split(' ')

    x = 0
    import string
    while x < len(result):
        for z in string.punctuation:
            result[x] = result[x].replace(z, '')            
        x = x + 1

    d = dict()
    for x in result:
        if(x.isalpha() == True):
            if (x not in d):
                d[x] = 1
            else:
                d[x] = d[x] + 1
                
    #불용어 제거
    f = open('stopwords.txt')
    s = f.read()
    s = s.splitlines()
    s.append('—')
    f.close()
    
    for x in s:
        if (d.get(x)):
            d.pop(x)
    
    return (d)
    print(d)
```

```python
class WebWordsFrequency(list):
    def __init__(self, *url): #생성자, url을 가변인수로 받음
        for e in url:
            self.append(e) #각각의 url을 추가함
    def addUrl(self, url): #addUrl 메소드
        self.append(url) #입력된 url을 추가함
    def removeUrl(self, url): #removeUrl 메소드
        if url in self: #입력된 url이 존재하면
            self.remove(url) #그 url을 삭제함
    def listUrls(self):
        for e in self: #모든 url 출력
            print (e)
    def getWordsFrequency(self):
        result = dict()
        for e in self: #각각의 url을 e라는 변수로 가져옴
            d = PrintHtml(e) #d = 각각 url 단어 빈도 사전
            for x in d: #빈도 사전의 원소
                if (x in result): #result에 있으면 x를 key로 d[x]만큼 value 추가
                    result[x] = result[x] + d[x]
                else: #result에 없으면 d[x]의 값이 value가 됨
                    result[x] = d[x]
        return result
    def getMaxFrequencyWords(self):
        print()
        WF = self.getWordsFrequency() #종합단어빈도
        result = list()
        if(WF != None) : 
            l = list(WF.values()) #종합 단어 빈도의 값
            M = max(l) #가장 큰 빈도수
            l = list(WF.items()) #종합단어빈도의 모든 item
            for e in l: #모든 아이템 중에서
                if(e[1] == M): #value의 값이 Max와 같으면
                    print(e[0]) #그 key(단어)를 출력하고
                    result.append(e[0]) #key(단어)를 result에 추가함
            return result
        else: #단어빈도사전에 아무것도 없을때
            print (None)
            return None
    def searchUrlByWord(self, word): #각각의 url 검색
        print()
        similarity = dict() #유사도 사전
        for e in self: #각각의 url에 대해 유사도 추출
            similarity[e] = PrintHtml(e).get(word, 0)
            
        M = max(similarity.values()) #Max값
        l = list(similarity.items()) #사전에서 item 추출
        result = list()
        for e in l: #각각의 item에 대해
            if(e[1] == M): #Max값과 같으면
                result.append(e[0]) #result에 추가
        print(result) #result 출력
        return result
```

```python
w1 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com', 'http://www.fox.com')
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com')
w3 = WebWordsFrequency()

w1.addUrl('http://github.com')
w3.addUrl('http://stackoverflow.com')

w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')

w1.getMaxFrequencyWords()
w3.getMaxFrequencyWords()
w1.searchUrlByWord("The")
w2.searchUrlByWord("CNN")
```

## 8. 위 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

    ◾ 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
    ◾ getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

        ◾ reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
        ◾ reverse 인자에 true를 넣으면 오름차순으로 정렬한다.

            >>> w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

            >>> w4.getWordsFrequency()
            [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

            >>> w4.getWordsFrequency(reverse=true)
            [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

## 9. 다음과 같은 코딩이 가능하도록 OrderedWebWordsFrequency 안에 반복자와 관련된 메소드를 추가하시오.

        >>> for i in w4:
        >>>       print(i)
        ('site, 12)
        ('science', 11)
        ('hello', 8)
        ('world', 2)
        ('program', 1)
        ('python', 1)

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self, *url): #생성자
        self.data = list()        
        WebWordsFrequency.__init__(self, *url) #WebWordsFrequency의 생성자를 불러옴
        self.result = dict() #반복자를 위한 변수
    def getWordsFrequency(self, reverse = False): #오버라이드
        print()
        e = 0
        self.result = dict()
        while e < len(self): #반복자에서 return되는 값이 바뀌어서 for in이 작동하지 않아 while로 대체함
            d = PrintHtml(self[e]) #각 url의 단어빈도 사전
            for x in d: #단어빈도 사전에서 각각의 값
                if (x in self.result): #result에 있으면 d[x]와 value를 합
                    self.result[x] = self.result[x] + d[x]
                else: #result에 없으면 d[x]가 value
                    self.result[x] = d[x]
            e = e + 1
        self.result = sorted(self.result.items(),key=lambda x:(x[1],x[0]), reverse = reverse) #정렬
        print(self.result)
        return self.result
    def __iter__(self): #반복자
        self.index = 0
        self.data = self.result #종합단어빈도사전
        return self
    def __next__(self):
        if self.index >= len(self.data): #종합단어빈도사전의 길이보다 길면 반복 중단
            raise StopIteration
        n = self.data[self.index]
        self.index += 1 #index 증가
        return n
```

```python
w4 = OrderedWebWordsFrequency('http://www.times.com', 'http://www.fox.com', 'http://github.com')

w4.getWordsFrequency()
w4.getWordsFrequency(reverse = True)

print()

for i in w4:
    print(i)
```
